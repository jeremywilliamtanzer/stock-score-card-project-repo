# API News exploration

## Exploring Polygon API for news to conduct eventually NLP, sentiment analysis

```python
#lib imports
import os
import langdetect
import requests
import pandas as pd
import numpy as np
import nltk
from gensim.models import Word2Vec
from datetime import date
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
NEWS_KEY = os.environ.get('NEWS_KEY')
```

```python
today = date.today()
today = today.strftime("%d-%m-%Y")
```

```python
today
```

```python
company_name = 'tesla'
country = 'us'
url = f'https://newsapi.org/v2/everything?q={company_name}&from={today}&sortBy=publishedAt&apiKey={NEWS_KEY}'
#url = f'https://newsapi.org/v2/everything?q={country}&from=2023-03-17&to=2023-03-18&sortBy=popularity&apiKey={NEWS_KEY}'
response = requests.get(url)
response = response.json()
response
```

```python
#article_content = response['articles'][0]['content']
#article_content

#all_contents = ''

#for article in response['articles']:
    #content = article['content']
    #all_contents += content

    
articles_content = []
for article in response['articles']:
    content = article['content']
    if content:
        try:
            if langdetect.detect(content) == 'en':
                articles_content.append(content)
        except:
            pass
```

```python
content_string = ''.join(articles_content)
content_string
```

```python
analyzer = SentimentIntensityAnalyzer()
score = analyzer.polarity_scores(content_string)
percentage_score = score['compound']*100
percentage_score = f'{percentage_score} %'
percentage_score = {'percentage_score':percentage_score}
percentage_score
```
