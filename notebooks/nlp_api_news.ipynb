```python
company = 'apple'
country = 'us'
#url = f'https://newsapi.org/v2/everything?q={company}&from=2023-03-17&sortBy=publishedAt&apiKey={news_key}'
url = f'https://newsapi.org/v2/everything?q={country}&from=2023-03-17&to=2023-03-18&sortBy=popularity&apiKey={news_key}'
response = requests.get(url)
response = response.json()
response
```
