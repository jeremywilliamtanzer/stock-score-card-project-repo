# getting financials from API ALPHAVANTAGE

```python
import pandas as pd
import requests
```

```python
alphavantage_key = "7QD6TN4TDCX63BZ6"
ticker = "aapl"
url = 'https://www.alphavantage.co/query?function=OVERVIEW&symbol=' + ticker + '&apikey=' + alphavantage_key
r = requests.get(url)
data = r.json()
```

```python
data.keys()
```

```python
# Selection of variables we want to use from this url (add/delete as needed)
keys_to_print = ['Name', 'Description', 'CIK', 'Industry', 'EPS', 
                 'ReturnOnAssetsTTM', 'EBITDA', 'PERatio', 'PEGRatio', 'BookValue', 
                 'ReturnOnEquityTTM', 'DividendPerShare', 'DividendYield', 
                 'AnalystTargetPrice', 'TrailingPE', 'ForwardPE']
```

```python
# Print the keys and values based on the list of keys
for key in keys_to_print:
    if key in data:
        value = data[key]
        print(key + ' : ' + value)
```
